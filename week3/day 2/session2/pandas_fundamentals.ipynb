```python
def rating_fn(r: float) -> str:
    if r < 8:
        return "bad"
        
    return "good"
```

```python
movies["rating_category"] = movies["rating"].apply(rating_fn)
```

```python
movies
```

```python
from typing import List
def has_comedy_fn(genres: List[str]) -> bool:
    return 'Comedy' in genres


movies[movies['genre'].str.split(',').apply(has_comedy_fn)]
```

```python
movies['genre'].str.split(',').apply(has_comedy_fn).sum()
```
